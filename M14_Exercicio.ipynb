## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

import seaborn as sns
import matplotlib.pyplot as plt

grafico = sns.barplot(data=data, y="cut", x="price", ci=None, palette="dark")
grafico.set(title='Preço do diamante por tipo de corte', ylabel='Corte', xlabel='Preço(USD)');
```

**Insight do gráfico 1**: O gráfico revela que o corte “Premium” tende a ter o preço médio mais alto, seguido pelo corte “Fair”. Isso pode indicar que esses cortes são mais valorizados no mercado de diamantes. Por outro lado, os cortes “Ideal” e “Good” têm preços médios mais baixos, sugerindo que podem ser mais comuns ou menos valorizados.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data2 = sns.load_dataset("flights")
data2.head()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar os dados para o mês de dezembro
dezembro = sns.load_dataset("flights").query("month == 'Dec'")

# Criar o gráfico de linhas
grafico = sns.lineplot(data=dezembro, x='year', y='passengers')
grafico.set(title='Número de passageiros em dezembro por ano', xlabel='Ano', ylabel='Número de Passageiros')

# Adicionar linhas de grade
plt.grid(True)

# Preencher a área abaixo da linha
plt.fill_between(dezembro['year'], dezembro['passengers'], color='skyblue', alpha=0.3)

plt.show()

```

**Insight do gráfico 2**: O gráfico mostra um aumento constante no número de passageiros em dezembro ao longo dos anos. Isso pode indicar que o setor de aviação estava crescendo durante esse período. Além disso, o fato de que o número de passageiros aumenta em dezembro de cada ano pode sugerir que dezembro é um mês popular para viagens aéreas, possivelmente devido às férias.
