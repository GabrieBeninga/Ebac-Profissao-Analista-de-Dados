## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = sns.boxplot(x=data["price"])
diamonds.set(title='Preços outliers de diamante', xlabel='Preço');
```

**Insight do gráfico 1**: Podemos ver que a média do preço dos diamantes é 2500 variando ente 1000 a 5000, e que os outiliers é sempre preços de 12000 em diante!

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 2

with sns.axes_style('whitegrid'):

  diamonds= sns.scatterplot(data=data, x="price", y="carat", hue="clarity", palette="dark")
  diamonds.set(title='Correlação entre preço e peso', xlabel='Preço', ylabel='Peso');
  diamonds.get_legend().set_title("Transparêcia");
```

**Insight do gráfico 2**: Há uma tendência geral de aumento do peso com o aumento do preço. Quanto maior, mais caro. E que O I1 é o diamante mais pesado do que outras transparência!

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3

cars = sns.histplot(data=data, x="weight", kde=True)

with sns.axes_style('whitegrid'):

  grafico = sns.histplot(data=data, x="weight", kde=True)
  grafico.set(title='Distribuição contínua aproximada do peso', xlabel='Peso', ylabel='Contagem');
```

**Insight do gráfico 3**: A distribuição do peso parece ser unimodal, com um pico em torno de um valor específico de peso (2000-2500). Isso indica que a maioria dos carros tem um peso em torno desse valor.
